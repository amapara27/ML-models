```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data
```

```python
test_data
```

```python
train_data["Sex_code"] = train_data["Sex"].astype("category").cat.codes
test_data["Sex_code"] = test_data["Sex"].astype("category").cat.codes
```

```python
train_data
```

```python
train_data.corr()["Survived"]
```

```python
train_data["Embarked_code"] = train_data["Embarked"].astype("category").cat.codes
test_data["Embarked_code"] = test_data["Embarked"].astype("category").cat.codes
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data["SibSp"] = np.log(train_data["SibSp"] + 1)
train_data["Fare"] = np.log(train_data["Fare"] + 1)
```

```python
train_data.corr()["Survived"]
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = train_data[["Sex_code", "Embarked_code"]]
y = train_data[["Survived"]]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
reg.fit(X_train, np.ravel(y_train))
```

```python
reg.score(X_train, y_train)
```

```python
t_predictors = test_data[["Sex_code", "Embarked_code"]]
```

```python
preds = reg.predict(t_predictors)
```

```python
test_data["Survived"] = preds
```

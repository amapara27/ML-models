```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data
```

```python
train_data = train_data.fillna(0)
test_data = test_data.fillna(0)
```

```python
train_data.shape
```

```python
train_data["HP_code"] = train_data["HomePlanet"].astype("category").cat.codes
test_data["HP_code"] = test_data["HomePlanet"].astype("category").cat.codes
```

```python
train_data["CS_code"] = train_data["CryoSleep"].astype("category").cat.codes
test_data["CS_code"] = test_data["CryoSleep"].astype("category").cat.codes
```

```python
train_data["Cabin_code"] = train_data["Cabin"].astype("category").cat.codes
test_data["Cabin_code"] = test_data["Cabin"].astype("category").cat.codes
```

```python
train_data["Destination_code"] = train_data["Destination"].astype("category").cat.codes
test_data["Destination_code"] = test_data["Destination"].astype("category").cat.codes
```

```python
train_data["VIP_code"] = train_data["VIP"].astype("category").cat.codes
test_data["VIP_code"] = test_data["VIP"].astype("category").cat.codes
```

```python
train_data = train_data.drop(['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP', 'Name'], axis=1)
test_data = test_data.drop(['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP', 'Name'], axis=1)
```

```python
train_data
```

```python
train_data.corr()["Transported"]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
X = train_data.drop(["Transported"], axis=1)
```

```python
y = train_data["Transported"]
```

```python
rf.fit(X, y)
```

```python
rf.score(X, y)
```

```python
test_data
```

```python
preds = rf.predict(test_data)
```

```python
test_data["Transported"] = preds 
```
